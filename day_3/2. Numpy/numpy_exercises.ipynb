### g) - i)

```python
#g. Create a 2d array with 1 on the border and 0 inside
N1 = 5
N2 = 10

# Naive way
g = np.zeros((N1,N2))
g[:,0] = 1
g[0,:] = 1
g[:,N2-1] = 1
g[N1-1,:] = 1

# Using padding function from https://www.geeksforgeeks.org/python/how-to-add-a-border-around-a-numpy-array/
# Although this ADDS a border around it instead of just replacing the one above
# Subtract two from both sides since pad adds that many elements in each axis
g_pad = np.pad(np.zeros((N1-2,N2-2)), pad_width=1, mode="constant", constant_values=1)
print(f" g): g_naive: \n {g} \n g_pad: \n {g_pad} \n\n")


# h. Create a 8x8 matrix and fill it with a checkerboard pattern
dim = 8
h = np.indices((dim,dim)).sum(axis=0) % 2
#dim = 9
#h = np.resize([1, 0], dim**2).reshape(dim, dim) # only fworks for odd dimensions
print(f" h): \n {h} \n")

#i. Create a checkerboard 8x8 matrix using the tile function
tile = np.array([[1, 0],[0, 1]]) # only works for even dimensions
i = np.tile(tile, (dim//2, dim//2))
print(f" i): \n {i} \n")
```
